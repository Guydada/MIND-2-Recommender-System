```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ti = pd.read_csv('../results/tfidf_scores_MindTF_IDF_train_size_49108.pkl -final.csv')
fg_tc = pd.read_csv("../results/tfidf_scores_MindTF_IDF_train_cols_['title', 'category']_size_49108.csv")
```

```python
df_ti
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (15, 10)
sns.boxplot(data=df_ti.drop(['behave_index'], axis=1))
# add labels
plt.xlabel('Score Metric')
plt.ylabel('nDCG Score')
plt.title('nDCG Score')
# add labels of mean and std on the boxplot on each box
plt.text(0.5, 0.9, 'Mean: ' + str(df_ti.drop(['behave_index'], axis=1).mean().round(2)),
         horizontalalignment='center',
         verticalalignment='center',
         transform=plt.gca().transAxes)
plt.show()
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (15, 10)
sns.boxplot(data=fg_tc.drop(['behave_index'], axis=1))
# add labels
plt.xlabel('Score Metric')
plt.ylabel('nDCG Score')
plt.title('nDCG Score')
# add labels of mean and std on the boxplot on each box
plt.text(0.5, 0.9, 'Mean: ' + str(df_ti.drop(['behave_index'], axis=1).mean().round(2)),
         horizontalalignment='center',
         verticalalignment='center',
         transform=plt.gca().transAxes)
plt.show()
```
